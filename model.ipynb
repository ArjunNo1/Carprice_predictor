```python
new_data=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
new_data.head()
```

```python
new_data['currentyear']=2022
new_data.head()
```

```python
new_data['No_of_yrs']=new_data['currentyear']-new_data['Year']
new_data.head()
```

```python
new_data.drop(['Year'],axis=1,inplace=True)
new_data.head()
```

```python
new_data.drop(['currentyear'],axis=1,inplace=True)
new_data.head()
```

```python
new_data=pd.get_dummies(new_data,drop_first=True)
new_data.head()
```

```python
new_data.corr()
```

```python
import seaborn as sns
sns.pairplot(new_data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))
sns.heatmap(new_data.corr(),annot=True,cmap='RdYlGn')
```

```python
new_data.columns
```

```python
X=new_data.iloc[:,1:]
y=new_data.iloc[:,0]
```

```python
# feature importance
from sklearn.ensemble import ExtraTreesRegressor
tree=ExtraTreesRegressor()
tree.fit(X,y)
```

```python
print(tree.feature_importances_)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
y_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
ran_forest=RandomForestRegressor()
```

```python
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]
max_features=['auto','sqrt']
max_depth=[int(x) for x in np.linspace(5,30,num=6)]
min_samples_split=[2,5,10,5,100]
min_samples_leaf=[1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid={'n_estimators':n_estimators,'max_features':max_features,'max_depth':max_depth,'min_samples_split':min_samples_split,'min_samples_leaf':min_samples_leaf}
```

```python
print(random_grid)
```

```python
rcv=RandomizedSearchCV(estimator=ran_forest,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
rcv.fit(X_train,y_train)
```

```python
predictor=rcv.predict(X_test)
predictor
```

```python
sns.distplot(y_test-predictor)
# it is normally distributed
```

```python
sns.scatterplot(y_test,predictor)
# It is linearly distributed
```

```python
import pickle
# open the file where you want to store the data

file=open('RandomForestModel.pkl','wb')
# dump the model to file
pickle.dump(rcv,file)
```
